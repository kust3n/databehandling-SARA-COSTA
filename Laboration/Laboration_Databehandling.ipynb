```python
sns.scatterplot(x='S.F.Ratio', y='Grad.Rate', hue='Elite', data=college)
plt.title("Student/Faculty Ratio versus Grad. Rate")
```

## Student/Faculty Ratio versus Grad. Rate
### Med denna graf ville jag undersöka skillnader mellan elit­skolor och icke-elit­skolor när det gäller antalet studenter i förhållande till antalet lärare samt examensfrekvensen. Min hypotes var att elit­skolor generellt har en högre examensfrekvens på grund av den hårda konkurrensen vid antagning.
### Som plotten visar så håller sig majoriteten av elitskolor på en högre procent av examensfrekvens. Men man ser även att elitskolor generellt har mindre antal studenter för varje lärare, detta kan tyda på att lägre antal studenter för varje lärare ökar examensfrekvensen, kanske då lärarna har mer tid för varje enskild elev och deras utveckling.

```python
sns.boxplot(x="Elite", y="perc.alumni", hue="Elite", data=college)
plt.title("Alumni donations by Elite")
```

## Alumni donations by Elite
### Här fokuserade jag mer på antal donationer som görs inom elitskolor och icke-elitskolor. Man ser en tydlig ökning bland elitskolor när det kommer till donationer som görs, något som stämde väl med min hypotes innan jag gjorde denna undersökningen

```python
sns.barplot(x="Elite", y="PhD", hue="Elite", data=college)
plt.title("Percent of Ph.D.s in Elite and non-Elite")
```

## Percent of Ph.D.s in Elite and non-Elite
### Jag vill undersöka och jämföra elitskolor och icke-elitskolor när det kommer till antal lärare som har en doktorsexamen (PhD), min hypotes var att elitskolor generellt har högre utbildade lärare vilket man tydligt ser på denna graf. Procentuellt har endast 70% av lärare inom icke-elitskolor en doktorsexamen, jämfört med elitskolor som når runt 90% av lärare. 

```python
sns.boxplot(x="Elite", y="Top10perc", hue="Elite", data=college)
plt.title("New students from top 10 in Elite and non-Elite")
```

## New students from top 10 in Elite and non-Elite
